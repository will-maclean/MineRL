```python
import os
from datetime import datetime
from pathlib import Path
import pickle
import numpy as np
import pandas as pd

from actions_utils import NULL_ACTION, decode_action
```

```python
__file__ = os.getcwd()

ROOT_PATH = Path(__file__).absolute().parent.parent.parent
SRC_PATH = ROOT_PATH.joinpath('src')
DATA_PATH = ROOT_PATH.joinpath('data')
ACTIONS_PATH = SRC_PATH.joinpath('actions')
```

```python
os.listdir(ACTIONS_PATH)
```

```python
def file_to_df(filename):
    file_path = ACTIONS_PATH.joinpath(f"{filename}.pickle")
    with open(file_path, 'rb') as f:
        d = pickle.load(f)

    return pd.DataFrame([decode_action(a) for a in d])
```
