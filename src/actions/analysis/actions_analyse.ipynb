```python
import os
from datetime import datetime
from pathlib import Path
import pickle
import numpy as np
import pandas as pd
import sys

__file__ = os.getcwd()

ROOT_PATH = Path(__file__).absolute().parent.parent.parent
SRC_PATH = ROOT_PATH.joinpath('src')
DATA_PATH = ROOT_PATH.joinpath('data')
ACTION_SET_PATH = DATA_PATH.joinpath('action_sets')
ACTIONS_PATH = SRC_PATH.joinpath('actions')

sys.path.append(str(ACTIONS_PATH))

from actions_utils import NULL_ACTION, decode_action

def file_to_df(filename):
    file_path = ACTION_SET_PATH.joinpath(f"{filename}.pickle")
    with open(file_path, 'rb') as f:
        d = pickle.load(f)

    return pd.DataFrame([decode_action(a) for a in d])
```
